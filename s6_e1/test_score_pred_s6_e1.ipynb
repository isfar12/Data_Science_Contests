```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../s6_e1/playground-series-s6e1/train.csv")
test = pd.read_csv("../s6_e1/playground-series-s6e1/test.csv")
```

```python
train.head()
```

```python
X=train.drop(columns=["id", "exam_score"])
y=train["exam_score"]
Test=test.drop(columns=["id"])
```

```python
print(train.columns.tolist())
```

```python
train.course.value_counts(), train.study_method.value_counts(), train.sleep_quality.value_counts(), train.facility_rating.value_counts(), train.exam_difficulty.value_counts()
```

```python
def internet_access_map(val):
    if val == "no":
        return 0
    elif val == "yes":
        return 1
    
def sleep_quality_map(val):
    if val == "poor":
        return 0
    elif val == "average":
        return 1
    elif val == "good":
        return 2
    
def facility_rating_map(val):
    if val == "low":
        return 0
    elif val == "medium":
        return 1
    elif val == "high":
        return 2

def exam_difficulty_map(val):
    if val == "easy":
        return 0
    elif val == "moderate":
        return 1
    elif val == "hard":
        return 2
```

```python
X['internet_access'] = X['internet_access'].map(internet_access_map)
X['sleep_quality'] = X['sleep_quality'].map(sleep_quality_map)
X['facility_rating'] = X['facility_rating'].map(facility_rating_map)
X['exam_difficulty'] = X['exam_difficulty'].map(exam_difficulty_map)

Test['internet_access'] = Test['internet_access'].map(internet_access_map)
Test['sleep_quality'] = Test['sleep_quality'].map(sleep_quality_map)
Test['facility_rating'] = Test['facility_rating'].map(facility_rating_map)
Test['exam_difficulty'] = Test['exam_difficulty'].map(exam_difficulty_map)
```

```python
X.head()
```

```python
X=pd.get_dummies(X, columns=['gender','course', 'study_method'], drop_first=True)
X.head()
```

```python
Test=pd.get_dummies(Test, columns=['gender','course', 'study_method'], drop_first=True)
Test.head()
```

```python
bool_columns = X.select_dtypes(include='bool').columns
X[bool_columns] = X[bool_columns].astype(int)

bool_columns_test = Test.select_dtypes(include='bool').columns
Test[bool_columns_test] = Test[bool_columns_test].astype(int)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
Test_scaled = scaler.transform(Test)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
Test_scaled = pd.DataFrame(Test_scaled, columns=Test.columns)
X_scaled.head()
```

### Torch Model for Score Prediction

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim
import torch.nn.functional as F

from sklearn.model_selection import train_test_split

# Set device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

X_train_tensor = torch.tensor(X_scaled.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y.values, dtype=torch.float32).view(-1, 1)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)


class ScorePredictionModel(nn.Module):
    def __init__(self, input_dim):
        super(ScorePredictionModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
input_dim = X_scaled.shape[1]
model = ScorePredictionModel(input_dim)
model.to(device)  # Move model to GPU
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 100

from tqdm import tqdm

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for inputs, targets in tqdm(train_loader):
        # Move data to GPU
        inputs = inputs.to(device)
        targets = targets.to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
    
    epoch_mse = running_loss / len(train_loader.dataset)
    epoch_rmse = np.sqrt(epoch_mse)  # Calculate RMSE from MSE
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch {epoch+1}/{epochs}, MSE: {epoch_mse:.4f}, RMSE: {epoch_rmse:.4f}')
```

```python
model.eval()
X_test_tensor = torch.tensor(Test_scaled.values, dtype=torch.float32)
X_test_tensor = X_test_tensor.to(device)  # Move test data to GPU

with torch.no_grad():
    test_predictions = model(X_test_tensor).cpu().numpy()  # Move predictions back to CPU
    
submission = pd.DataFrame()
submission['id'] = test['id']
submission['exam_score'] = test_predictions
submission.to_csv('submission_torch_model.csv', index=False)
```

### LightGBM Model for Score Prediction

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from lightgbm import LGBMRegressor

# Create KFold cross-validation (stratified is typically for classification, so using KFold for regression)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store scores for each fold
fold_scores = []
rmse_scores = []

print("Starting K-Fold Cross Validation...\n")

for fold, (train_idx, val_idx) in enumerate(kf.split(X_scaled), 1):
    # Split data
    X_train_fold = X_scaled.iloc[train_idx]
    X_val_fold = X_scaled.iloc[val_idx]
    y_train_fold = y.iloc[train_idx]
    y_val_fold = y.iloc[val_idx]
    
    # Train model
    model_fold = LGBMRegressor(n_estimators=1000, learning_rate=0.05, random_state=42)
    model_fold.fit(X_train_fold, y_train_fold)
    
    # Predict
    y_pred = model_fold.predict(X_val_fold)
    
    # Calculate metrics
    r2 = r2_score(y_val_fold, y_pred)
    rmse = np.sqrt(mean_squared_error(y_val_fold, y_pred))
    
    fold_scores.append(r2)
    rmse_scores.append(rmse)
    
    print(f"Fold {fold}:")
    print(f"  R Score: {r2:.4f}")
    print(f"  RMSE: {rmse:.4f}\n")

print(f"Average R Score: {np.mean(fold_scores):.4f} (+/- {np.std(fold_scores):.4f})")
print(f"Average RMSE: {np.mean(rmse_scores):.4f} (+/- {np.std(rmse_scores):.4f})")
```
