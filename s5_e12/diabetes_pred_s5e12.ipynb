```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
numerical_cols.remove("diagnosed_diabetes")
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer

transformer=ColumnTransformer(
    transformers=[
        ('cat', OrdinalEncoder(),categorical_cols),
        ('num', StandardScaler(), numerical_cols),
    ]
)
X_train_scaled=transformer.fit_transform(Train_X)
X_test_scaled=transformer.transform(test)
```

```python
X_train_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train_scaled, Train_Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42,)

model.fit(X_train, y_train)
y_pred = model.predict(X_val)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print("Accuracy:", accuracy_score(y_val, y_pred))
print(confusion_matrix(y_val, y_pred))
print(classification_report(y_val, y_pred))
```

```python
import lightgbm as lgb

lightgbm_model = lgb.LGBMClassifier(n_estimators=100, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score,KFold, StratifiedKFold

kf=StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores=cross_val_score(lightgbm_model, X_train_scaled, Train_Y, cv=kf, scoring='accuracy')
print("Cross-validation scores:", cv_scores)
```

```python
from sklearn.metrics import roc_auc_score


cv_results={}

param_grid = [
    {'num_leaves': 31, 'max_depth': -1},
    {'num_leaves': 63, 'max_depth': -1},
    {'num_leaves': 31, 'max_depth': 10},
    {'num_leaves': 63, 'max_depth': 10},
]

base_params = {
    'objective': 'binary',
    'metric': 'auc',
    'learning_rate': 0.05,
    'verbosity': -1
}
for params in param_grid:
    fold_aucs = []
    
    for tr_idx, val_idx in kf.split(Train_X, Train_Y):
        X_tr,X_val = Train_X.iloc[tr_idx], Train_X.iloc[val_idx]
        y_tr, y_val = Train_Y.iloc[tr_idx], Train_Y.iloc[val_idx]
        
        model = lgb.LGBMClassifier(n_estimators=100, random_state=42, **base_params)
        model.fit(transformer.transform(X_tr), y_tr, eval_set=[(transformer.transform(X_val), y_val)], callbacks=[lgb.early_stopping(10)])
        
        preds = model.predict_proba(transformer.transform(X_val))[:, 1]
        auc = roc_auc_score(y_val, preds)
        fold_aucs.append(auc)
    
    mean_auc = np.mean(fold_aucs)
    cv_results[str(params)] = mean_auc
    
print("Cross-validation results:", cv_results)
```

```python
best_params = max(cv_results, key=cv_results.get)
print("Best parameters:", best_params)
```

```python
best_params = eval(best_params)

final_model = lgb.LGBMClassifier(
    **base_params,
    **best_params,
    n_estimators=5000
)

final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(transformer.transform(test))

submission_data = pd.DataFrame()
test=pd.read_csv("../s5_e12/playground-series-s5e12/test.csv")
submission_data['id'] = test['id']
submission_data['diagnosed_diabetes'] = y_pred
submission_data.to_csv("submission.csv", index=False)
```

```python
X_train_tensor = torch.FloatTensor(X_train)
y_train_tensor = torch.LongTensor(y_train.values)

X_val_tensor = torch.FloatTensor(X_val)
y_val_tensor = torch.LongTensor(y_val.values)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
```

```python
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
class DiabetesModel(nn.Module):
    def __init__(self, input_dim):
        super(DiabetesModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2=nn.Linear(64,32)
        self.fc3 = nn.Linear(32, 2)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)
        
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x
```

```python
import torchsummary
input_dim =X_train_tensor.shape[1]
model = DiabetesModel(input_dim)
torchsummary.summary(model, (input_dim,))
```

```python
epochs = 10
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
import tqdm

for epoch in tqdm.tqdm(range(epochs)):
    for inputs_batch, labels_batch in train_loader:
        
        optimizer.zero_grad()
        inputs_batch, labels_batch = inputs_batch.to(device), labels_batch.to(device)
        
        outputs =model(inputs_batch)
        loss = criterion(outputs, labels_batch)
        loss.backward()
        
        optimizer.step()
        
    if (epoch+1) % 10 ==0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')   
        
```

```python
model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    for inputs_batch, labels_batch in val_loader:
        inputs_batch, labels_batch = inputs_batch.to(device), labels_batch.to(device)
        outputs = model(inputs_batch)
        _, predicted = torch.max(outputs.data, 1)
        total += labels_batch.size(0)
        correct += (predicted == labels_batch).sum().item()
        
    print(f'Validation Accuracy: {100 * correct / total:.2f}%')
```
